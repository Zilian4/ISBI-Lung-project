```python
import pandas as pd
```

```python
data = pd.read_csv(r"F:\Myproject\ISBI-Lung-project\radiomics_features.csv")
selected_colums = data.columns[23:].to_list()
selected_colums.append('Image')
data = data[selected_colums]
```

#Read label

```python
label1 = pd.read_excel(r"F:\Myproject\ISBI-Lung-project\Covid_labels.xlsx")
label1["Image"] = label1["Image"].apply(lambda x: 'U_'+str(x)+'.nii.gz')
label2 = pd.read_excel(r"F:\Myproject\ISBI-Lung-project\i_labels.xlsx")
label2["Image"] = label2["Image"].apply(lambda x: 'T_'+str(x)+'.nii.gz')
label = pd.concat([label1, label2], ignore_index=True)
```

```python
data = pd.merge(data, label, on='Image',how='left')
data = data.dropna()
```

```python
data.to_csv('Lung_data.csv', index=False)
```
